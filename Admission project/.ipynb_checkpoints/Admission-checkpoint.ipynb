# Graduate admission Project

## Data extraction 

#### Context

This dataset is created for prediction of Graduate Admissions. This project has a fairly small dataset. It only has 500 observations. This data set has the information on the GRE,TOEFL,CGPA and other details of students seeking post graduation admission at Universities.

****

#### Dataset

This dataset is created for prediction of Graduate Admissions. The dataset contains several parameters which are considered important during the application for Masters Programs.

I will use the [Admission Data Set from Kaggle](https://www.kaggle.com/mohansacharya/graduate-admissions/home).

Features in the dataset:

- GRE Scores (290 to 340)

- TOEFL Scores (92 to 120)

- University Rating (1 to 5)

- Statement of Purpose (1 to 5)

- Letter of Recommendation Strength (1 to 5)

- Undergraduate CGPA (6.8 to 9.92)

- Research Experience (0 or 1)

- Chance of Admit (0.34 to 0.97)

```python
admission = pd.read_csv("Admission_Predict.csv", sep='\s*,\s*')
```

## Data cleanup

#### Treating / testing missing values

```python
admission.isnull().sum()
```

#### Drop inrelevant columns

```python
admission.drop('Serial No.', axis=1, inplace=True)
```

#### Add a new column

I assume that students with 0.8 chance of admission have secured admission to the university. I create another column in our dataset named Admit_class. When the the value is 1, the Chance of admission >0.8 and the value is 0 the Chance of admission <0.8.

```python
def Admit_class(row):
    if row > 0.8:
        return 1
    else :
        return 0
    
admission['Admit_class'] = admission["Chance of Admit"].apply(Admit_class)
```

## Predictive analytics - Linear regression

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
X = admission.drop(["Admit_class","Chance of Admit"], axis=1)
y = admission["Chance of Admit"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.2,random_state=101)

reg = linear_model.LinearRegression()
reg.fit(X_train,y_train)
```

```python
print("Intercept: {}".format(reg.intercept_))
```

```python
coeff_df = pd.DataFrame(reg.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

```python
from sklearn.metrics import r2_score

pred = reg.predict(X_test)
print("Mean Squared Error: {}".format(mean_squared_error(y_test, pred)))

coefficient_of_dermination = r2_score(y_test, pred)
print("R2: {}".format(coefficient_of_dermination))
```

## Predictive analytics - Linear regression (Standardization)

```python
from sklearn.preprocessing import StandardScaler
```

```python
X1 = admission.drop(["Admit_class","Chance of Admit"], axis=1)
y1 = admission["Chance of Admit"]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(X1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(scaled_features, y1, test_size=0.2,random_state=101)
```

```python
reg1 = linear_model.LinearRegression()
reg1.fit(X_train,y_train)
```

```python
print("Intercept: {}".format(reg1.intercept_))
```

```python
coeff_df1 = pd.DataFrame(reg1.coef_, X.columns, columns=['Coefficient'])
coeff_df1
```

```python
pred1 = reg1.predict(X_test)
print("Mean Squared Error: ", mean_squared_error(y_test, pred1))

coefficient_of_dermination1 = r2_score(y_test, pred1)
print("R2: {}".format(coefficient_of_dermination1))
```

## Predictive analytics - Random Forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=101)

rfr = RandomForestRegressor(n_estimators = 100, random_state = 42)
rfr.fit(X_train,y_train)
```

```python
pred2 = rfr.predict(X_test)
print("Mean Squared Error: {}".format(mean_squared_error(y_test, pred2)))

coefficient_of_dermination = r2_score(y_test, pred2)
print("R2: {}".format(coefficient_of_dermination))
```

## Predictive analytics - Random Forest regressor (Tune the parameters)

```python
from sklearn.model_selection import GridSearchCV

param_grid = [{'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6]}]

forest_reg = RandomForestRegressor()

# train across 5 folds, that's a total of (12+6)*5=90 rounds of training 
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring="r2")

grid_search.fit(X_train, y_train)
```

```python
print("The best values for paramaters:", grid_search.best_params_)
print("The R2:", round(grid_search.best_score_,2))
```

Notice in the above models:

**Linear regression** would be the best model to make the prediction because it has the higher **R2** and smaller **mean squared error**.

## Predictive analytics - Logistic regression

```python
X2 = admission.drop(["Chance of Admit","Admit_class"], axis=1)
y2 = admission["Admit_class"]
```

```python
from sklearn.linear_model import LogisticRegression

X_train,X_test,y_train,y_test = train_test_split(X2, y2, test_size=0.2, random_state=101)
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
pred2 = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, pred2))
print(classification_report(pred2, y_test))
```

```python
sns.heatmap(confusion_matrix(y_test, pred2), annot = True, linewidths=0.05, cmap="magma")
plt.title("Test for Test Dataset")
plt.xlabel("predicted y values")
plt.ylabel("real y values")
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score

print("precision_score: ", precision_score(y_test, pred2))
print("recall_score: ", recall_score(y_test, pred2))
```
